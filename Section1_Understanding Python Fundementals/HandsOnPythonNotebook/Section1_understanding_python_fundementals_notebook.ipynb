### break / continue

```python
values = [5, 0, 2, 1]
```

```python
print(values)
```

```python
results = []
for v in values:
    if v == 0:
        continue  # skip divide by zero
    r = 10 / v
    results.append(r)
    if r < 3:
        break  # stop early
```

```python
print(results)
```

## 07 — Classes

### Class + objects

```python
class Wallet:
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError('Deposit must be positive')
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError('Withdraw must be positive')
        if amount > self.balance:
            raise ValueError('Insufficient funds')
        self.balance -= amount
        return self.balance

    def __repr__(self):
        return f'Wallet(owner={self.owner!r}, balance={self.balance:.2f})'
```

```python
w = Wallet('Michael', 75)
```

```python
print(w)
```

```python
w.deposit(25)
```

```python
print(w.balance)
```

```python
w.withdraw(40)
```

```python
print(w.balance)
```

### Inheritance

```python
class SavingsWallet(Wallet):
    def apply_interest(self, rate):
        if rate < 0:
            raise ValueError('Rate cannot be negative')
        self.balance *= (1 + rate)
        return self.balance
```

```python
sw = SavingsWallet('Jessica', 120)
```

```python
print(sw)
```

```python
sw.apply_interest(0.05)
```

```python
print(sw.balance)
```
